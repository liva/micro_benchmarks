```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("out", header=None, index_col=0)

plt.figure(figsize=(60, 50), dpi=50)
cnt = 100
for num in range(4096):
    plt.plot(np.full(cnt, num), data.iloc[num,0:cnt], '.')
plt.ylim([0,5000])
plt.show()
```
